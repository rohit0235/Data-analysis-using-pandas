```python
def t_b(x):
  if x<5:
    return "D"
  elif x>=5 and x<20:
    return "C"
  elif x>=20 and x<50:
    return "B"
  else:
    return "A"
```

```python
df['tenure_b']=df['tenure'].apply(t_b)
```

```python

df['tenure'].apply(lambda x: "D" if x<5 else ("C" if x>=5 and x<20 else ("B" if x>=20 and x<50 else "A")))
```
